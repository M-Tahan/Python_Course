#### 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(1, 11)
y = 2 * x + 5
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x, y)
plt.show()
```

#### 2. Write a Python program to plot two or more lines on same plot with suitable legends of each line.


```python
x = np.arange(1, 11)
x1 = np.arange(1, 11)
y = 2 * x + 5
y1 = 3 * x + 10
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x, y)
plt.plot(x1, y1)
plt.show()
```

#### 3. Write a Python program to plot two or more lines with legends, different widths and colors.



```python
x = np.arange(1, 11)
x1 = np.arange(1, 11)
y = 2 * x + 5
y1 = 3 * x + 10
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.plot(x, y, label="line 1", color="green")
plt.plot(x1, y1, label="line 2", color="red")
plt.legend()
plt.show()
```

#### 4. Write a Python program to plot two or more lines with different styles.


```python
plt.plot(x1, y1, label="line 2", color="red", linestyle="dashed")
plt.plot(x, y, label="line 1", color="green")
plt.legend()
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.show()
```

#### 5. Write a Python program to plot two or more lines and set the line markers.


```python
plt.plot(x1, y1, label="line 2", color="red", linestyle="dashed", marker="o")
plt.plot(x, y, label="line 1", color="green", marker="*")
plt.legend()
```

#### 6. Write a Python programming to display a bar chart of the popularity of programming Languages.
    Sample data:
    Programming languages: Java, Python, PHP, JavaScript, C#, C++
    Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7



```python
x = ["Java" , "Python" , "PHP" , "JavaScript" , "C#" , "C++"]
y = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.bar(x, y)
plt.xlabel("Language")
plt.ylabel("Popularity")
plt.show()
```

#### 8. Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.
    Sample Data:
    Means (men) = (22, 30, 35, 35, 26)
    Means (women) = (25, 32, 30, 35, 29)


```python
Means_men= (22, 30, 35, 35, 26)
Means_women = (25, 32, 30, 35, 29)
x = ["men1" , "men2" , "men3" , "men4" , "men5" ]
x1 = ["women1" , "women2" , "women3" , "women4" , "women5"]
plt.bar(x, Means_men)
plt.bar(x1, Means_women)
plt.xlabel("Groups")
plt.ylabel("Scores")
plt.show()
```

#### 9. Write a Python programming to create a pie chart of the popularity of programming Languages.



```python
x = ["Java" , "Python" , "PHP" , "JavaScript" , "C#" , "C++"]
y = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
plt.pie(y, labels=x)
plt.show()
```
