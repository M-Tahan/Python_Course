```python
a = np.ones((10,10))
a[1:-1,1:-1] = 0
print(a)
```

### 12. Write a NumPy program to create a 5x5 zero matrix with elements on the main diagonal equal to 1, 2,3, 4, 5.

```python
a = np.zeros((5,5))
a[np.arange(5),np.arange(5)] = [1,2,3,4,5]
print(a)
```
